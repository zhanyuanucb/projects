```python
import os, shutil, re
# Generate training, validation, and testing data folders
original_dataset = 'CarLogos/images'
base_dir = "data"
os.mkdir(base_dir)

train_dir = os.path.join(base_dir, "train")
os.mkdir(train_dir)
validation_dir = os.path.join(base_dir, "validation")
os.mkdir(validation_dir)
test_dir = os.path.join(base_dir, "test")
os.mkdir(test_dir)

# Get all 40 brands
all_images = os.listdir(original_dataset)
all_brands = set()
for file in all_images:
    all_brands.add(re.findall('[A-Za-z ]+', file)[0])
assert len(all_brands) == 40

data_folder = [train_dir, validation_dir, test_dir]
for folder in data_folder:
    for brand in all_brands:
        brand_dir = os.path.join(folder, brand)
        os.mkdir(brand_dir)

# Sanity check
for folder in data_folder:
    temp = os.listdir(folder)
    assert (len(temp) == 40)
```

```python
# Distribute images into training, validation, and testing folders

# Create a dictionary whose keys are car brands and values are corresponding images
brand_dict = {}
for brand in all_brands:
    brand_dict[brand] = []
for image in all_images:
    brand = re.findall("[A-Za-z ]+", image)[0]
    brand_dict[brand].append(image)

# Copy images into data folders
for brand, image_list in brand_dict.items():
    # train:valid:test = 80:10:10
    train_data = image_list[:round(len(image_list)*0.8)]
    valid_data = image_list[round(len(image_list)*0.8):round(len(image_list)*0.9)]
    test_data = image_list[round(len(image_list)*0.9):]
    
    train_dst = os.path.join(train_dir, brand)
    valid_dst = os.path.join(validation_dir, brand)
    test_dst = os.path.join(test_dir, brand)
    
    for image in train_data:
        src = os.path.join(original_dataset, image)
        dst = os.path.join(train_dst, image)
        shutil.copyfile(src, dst)
    for image in valid_data:
        src = os.path.join(original_dataset, image)
        dst = os.path.join(valid_dst, image)
        shutil.copyfile(src, dst)
    for image in test_data:
        src = os.path.join(original_dataset, image)
        dst = os.path.join(test_dst, image)
        shutil.copyfile(src, dst)
```

```python
# Instantiating convet for classification
from keras import layers, models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                       input_shape=(50, 50, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Dropout(0.5))
model.add(layers.Flatten())
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dense(40, activation='sigmoid'))
```

```python
# Choose loss function and optimizer
from keras.optimizers import SGD
sgd = SGD(lr = 0.005, decay = 1e-6, momentum = 0.95, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['categorical_accuracy'])
model.summary()
```

```python
# Data preprocessing
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_dir,
                                                   target_size=(50, 50),
                                                   batch_size=128,
                                                   class_mode='categorical')
validation_generator = test_datagen.flow_from_directory(validation_dir,
                                                       target_size=(50, 50),
                                                       batch_size=128,
                                                       class_mode='categorical')
```

```python
history = model.fit_generator(train_generator,
                             steps_per_epoch=100,
                             epochs=120,
                             validation_data=validation_generator,
                             validation_steps=50,
                             workers=4)
```

```python
# Curves of loss and accuracy during training
os.mkdir("output")
import matplotlib.pyplot as plt
acc = history.history['categorical_accuracy']
val_acc = history.history['val_categorical_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()
plt.savefig("output/diagnotic_plot.png")
plt.show()
```

```python
final_test_datagen = ImageDataGenerator(rescale=1./255)

final_test_generator = train_datagen.flow_from_directory(test_dir,
                                                         target_size=(50, 50),
                                                         batch_size=128,
                                                         class_mode='categorical')
scores = model.evaluate_generator(final_test_generator)
print("Final accuracy test: %.2f%%" % (scores[1]*100))
```
